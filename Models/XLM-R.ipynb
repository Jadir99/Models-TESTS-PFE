```python
from transformers import pipeline

ner = pipeline(
    "ner",
    model="Davlan/xlm-roberta-base-ner-hrl",
    aggregation_strategy="simple"
)

```

```python
text = """
Ingénieur Data avec expérience en Python, SQL et Machine Learning.
Diplômé de ENSA Safi.
"""

entities = ner(text)

for entity in entities:
    print(entity)
```
